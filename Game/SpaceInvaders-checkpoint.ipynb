```python
import pygame
import sys
import random
import math
from pygame import mixer

#initialize
pygame.init()

#create screen
screen = pygame.display.set_mode((800,600))

#Title and icon
pygame.display.set_caption("Space Invaders")
                           
#Background Image
backImg = pygame.image.load('Downloads/Game/background.png')

#score
score = 0
font = pygame.font.Font('freesansbold.ttf',32)
tX = 10
tY = 10

def Score(x,y):
    sc = font.render("Score: "+ str(score),True,(255,255,255))
    screen.blit(sc,(x,y))

#game over
over_font = pygame.font.Font('freesansbold.ttf',64)

def game_over_txt():
    go = over_font.render("GAME OVER",True,(255,255,255))
    screen.blit(go,(200,250))

#player
playerImg = pygame.image.load('Downloads/Game/si.png')
pX = 350
pY = 480
pX_cng = 0

#enemy
enemyImg = []
eX = []
eY = []
eX_cng = []
eY_cng = []
no_of_enemy = 4

for i in range(no_of_enemy):
    enemyImg.append(pygame.image.load('Downloads/Game/menemy.png'))
    eX.append(random.randint(0,736))
    eY.append(random.randint(50,100))
    eX_cng.append(3)
    eY_cng.append(30)

bnX = random.randint(0,763)
bnY = 10
bImg = pygame.image.load('Downloads/Game/star.png')


#bullet
#read - out of screen ; fire - on the screen
bulletImg = pygame.image.load('Downloads/Game/bullet.png')
bX = 0
bY = 480
bX_cng = 0
bY_cng = 10
b_state = "ready"

def Player(x,y):
    screen.blit(playerImg,(x,y))

def Enemy(x,y,i):
    screen.blit(enemyImg[i],(x,y))
    
def MEnemy(x,y):
    screen.blit(meImg,(x,y))
    
def Bonus(x,y):
        screen.blit(bImg,(x,y))
 
def fire_bullet(x,y):
    global b_state
    b_state = "fire"
    screen.blit(bulletImg,(x + 16 , y + 10))

def isCollision(eX,eY,bX,bY):
    dist = (math.sqrt(math.pow(eX-bX,2) + math.pow(eY-bY,2)))
    if dist <= 27:
        return True
    else:
        return False

# sound
mixer.music.load('Downloads/Game/background.wav')
mixer.music.play(-1)

state = "play"

#game loop

while True:
    # RGB
    screen.fill((0,0,0))

    screen.blit(backImg,(0,0))

    Score(tX,tY)
 
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

        #check if key is pressed
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                pX_cng -= 4
            if event.key == pygame.K_RIGHT:
                pX_cng += 4
            if event.key == pygame.K_SPACE:
                if b_state is "ready":
                    b_s = mixer.Sound('Downloads/Game/laser.wav')
                    b_s.play()
                    bX = pX
                    fire_bullet(bX,bY)
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                pX_cng = 0

    pX += pX_cng 
    if pX <= 0:
        pX = 0
    elif pX >= 736:
        pX = 736

    for i in range(no_of_enemy):

        #game over
        if isCollision(eX[i],eY[i],pX,pY):
            for j in range(no_of_enemy):
                eY[j] = 20000
                game_over_txt()
            break

        eX[i] += eX_cng[i] 
        if eX[i] <= 0:
            eX_cng[i] = 3
            eY[i] += eY_cng[i]
        elif eX[i] >= 736:
            eX_cng[i] = -3
            eY[i] += eY_cng[i]
        if isCollision(eX[i],eY[i],bX,bY):
            e_s = mixer.Sound('Downloads/Game/explosion.wav')
            e_s.play()
            score += 1
            b_state = "ready"
            bY = pY
            eX[i] = random.randint(0,736)
            eY[i] = random.randint(50,400)
        Enemy(eX[i],eY[i],i)
        
        bnY += 1
        Bonus(bnX,bnY)
        if isCollision(bnX,bnY,pX,pY):
            score += 2
            bnX = random.randint(0,763)
            bnY = 5
        if isCollision(bnX,bnY,bX,bY):
            score -= 1
            bnX = random.randint(0,763)
            bnY = 0

        if bnY >= 600:
            bnX = random.randint(0,763)
            bnY = 0
        
    if bY <= 0:
        b_state = "ready"
        bY = pY
    if b_state is "fire":
        fire_bullet(bX,bY)
        bY -= bY_cng
    
    Player(pX,pY)
                          
    pygame.display.update()
```
